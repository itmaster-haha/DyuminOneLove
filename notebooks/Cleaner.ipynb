```python
import pandas as pd
```

```python
df = pd.read_csv('merged_all_sources.csv', sep=';')
df.head()
```

```python
df.info(show_counts=True)
```

```python
cleaner_df = df.dropna(axis=1, how='all')
cleaner_df.head()
```

```python
cleaner_df = cleaner_df.drop('Fare', axis = 1)
cleaner_df = cleaner_df.drop('Baggage', axis = 1)
cleaner_df = cleaner_df.drop('Meal', axis = 1)
```

```python
print(cleaner_df[['FirstName', 'LastName', 'SecondName']][cleaner_df['SecondName'].notnull()])
```

```python
def standartize(name):
  if type(name) != str:
    return name
  name = name.upper()
  name = name.replace('Y', 'I')
  return name
```

```python
cleaner_df[['FirstName', 'LastName', 'SecondName']] = cleaner_df[['FirstName', 'LastName', 'SecondName']].applymap(standartize)
cleaner_df[['FirstName', 'LastName', 'SecondName']]
```

```python
def standartize_city(city):
  if type(city) != str:
    return city
  city = city.upper()
  return city
```

```python
cleaner_df[['ArrivalCity', 'DepartCity']] = cleaner_df[['ArrivalCity', 'DepartCity']].applymap(standartize_city)
cleaner_df[['ArrivalCity', 'DepartCity']]
```

```python
cleaner_df.info()
```

```python
def standartize_date(date):
  if type(date) != int:
    sdate = str(date)
    if type(date) == str:
      if sdate.find('/') != -1:
        return sdate[3:5] + '.' + sdate[:2] + '.' + sdate[-4:]
  sdate = '0' * (date < 10000000) + str(date)
  if date % 10000 > 1914:
    return sdate[:2] + '.' + sdate[2:4] + '.' + sdate[4:]
  else:
    return sdate[6:] + '.' + sdate[4:6] + '.' + sdate[:4]
```

```python
def standartize_birth(date):
  if type(date) != str:
    return date
  if date.find('/') != -1:
    return date[3:5] + '.' + date[:2] + '.' + date[-4:]
  return date
```

```python
cleaner_df[['DepartDate']] = cleaner_df[['DepartDate']].map(standartize_date)
```

```python
cleaner_df[['PassengerBirthDate']] = cleaner_df[['PassengerBirthDate']].map(standartize_birth)
```
